```python
import pandas as pd
df = pd.read_excel("Employees.xlsx",'Employees')
df
```

```python
from sklearn.cluster import KMeans #Categories known and < 10000

from sklearn.model_selection import train_test_split
X = df[['Salary']]
y = df[['Role']]

kmeans = KMeans(n_clusters=4,max_iter=10)#Find 4 clusters that data is arround
kmeans.fit(X)
print (kmeans.cluster_centers_)
print ('Employee with salary 9000 related  cluster is : ',kmeans.predict([[9000]]))#prediction for 9000
print ('Employee with salary 20000 related  cluster is : ',kmeans.predict([[20000]]))#prediction for 20000
print ('Employee with salary 25000 related  cluster is : ',kmeans.predict([[25000]]))#prediction for 25000
```

```python
from sklearn.cluster import MiniBatchKMeans #Categories known and > 10000

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=0)

kmeans = MiniBatchKMeans(n_clusters=4, max_iter=10)#Find 4 clusters that data is arround
kmeans.fit(X)
print (kmeans.cluster_centers_)
print ('Employee with salary 9000 related  cluster is : ',kmeans.predict([[9000]]))#prediction for 9000
print ('Employee with salary 20000 related  cluster is : ',kmeans.predict([[20000]]))#prediction for 20000
print ('Employee with salary 25000 related  cluster is : ',kmeans.predict([[25000]]))#prediction for 25000
```

```python
from sklearn.cluster import MeanShift   #dataset > 10000 and Categories unknown


kmeans = MeanShift(max_iter=10)#Find 4 clusters that data is arround
kmeans.fit(X)
print (kmeans.cluster_centers_)
print ('Employee with salary 9000 related  cluster is : ',kmeans.predict([[9000]]))#prediction for 9000
print ('Employee with salary 20000 related  cluster is : ',kmeans.predict([[20000]]))#prediction for 20000
print ('Employee with salary 25000 related  cluster is : ',kmeans.predict([[25000]]))#prediction for 25000
```
